```python
from utils.plot_utils import read_auto_labels_and_gold_labels
from utils.plot_utils import extract_labels
from utils.plot_utils import cm_analysis_plot
from utils.plot_utils import no_fact_to_off_topic
```

```python
auto_path = "./automatic_labels.jsonl"
gold_path = "../../gpt-neo/gen_outputs/labeled_samples/joined_labels.jsonl"
```

# Vanilla Evaluation

```python
auto_gen, gold_gen = read_auto_labels_and_gold_labels(auto_path, gold_path)

labels_gold = extract_labels(gold_gen)
labels_auto = extract_labels(auto_gen)

assert len(labels_gold) == len(labels_auto), f"Len auto: {len(labels_auto)}, Len gold: {len(labels_gold)}"
```

```python
count = 0
for gold, auto in zip(gold_gen, auto_gen):
    len_gold = len(gold["single_sentences"])
    len_auto = len(auto["single_sentences"])

    if len_gold != len_auto:
        print(count)

    count += 1
```

```python
labels = ["true", "false", "not_enough_evidence", "off_topic", "no_fact"]
```

```python
cm_analysis_plot(y_true=labels_gold , y_pred=labels_auto, filename="./cm.png", labels=labels)
```

```python
new_gold, new_auto = no_fact_to_off_topic(labels_gold, labels_auto)

new_labels = ["true", "false", "not_enough_evidence", "off_topic"]
```

```python
cm_analysis_plot(y_true=new_gold , y_pred=new_auto, filename="./cm_off_topic.png", labels=new_labels)
```
